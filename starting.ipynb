```python
import matplotlib.pyplot as plt
import seaborn as sns

import bokeh as bk

from plotnine import *
import plotnine.options as pno

import altair as alt

import plotly.express as px
import plotly.graph_objects as go


import pandas as pd
import numpy as np

# Installs with pip3
```

```python
dataset = pd.read_csv("data.csv")
```

Why are we getting rid of matplotlib and seaborn?

Convoluted, ugly, and too difficult.

If we start with very simple graphs, it's not too awful.

```python
n, bins, patches = plt.hist(dataset.acousticness, 50, facecolor='g', alpha=0.75)
plt.xlabel('Acousticness')
plt.title('Histogram of Acousticness')
plt.grid(True)
plt.show()
```

```python
with sns.axes_style("whitegrid"):
    sns.histplot(data=dataset, x="acousticness",  binwidth=.02).set_title("Histogram of Acousticness")
```

```python
from bokeh.io import show, output_notebook
from bokeh.plotting import figure

hist, edges = np.histogram(dataset.acousticness, bins=50)

p = figure(title="Histogram of Acousticness", y_axis_label='Count', x_axis_label='Acousticness', width=700, height = 300)

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:])
output_notebook() 
show(p)
```

```python
source = dataset.sample(axis = 0, n=4000)

alt.Chart(source).mark_bar().encode(
    alt.X("acousticness", bin=True),
    y='count()',
).properties(
    title='Histogram of Acousticness'
)
```

```python

pno.dpi = (100)
pno.figure_size = (8,3)
ggplot(data=dataset, mapping=aes(x='acousticness')) + theme_bw() + geom_histogram(color='white', bins=50) + labs(title = "Histogram of Acousticness")
```
